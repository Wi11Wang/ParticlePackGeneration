```python
def get_cmap(n, cmap='gist_rainbow'):
    """
    Create a custom color map that plots n colors, while leaving the background black
    """
    original_cmap = mpl.colormaps[cmap]
    colors = np.linspace(0, 1, n - 1)
    cmap_colors = original_cmap(colors)
    black = np.array([[0, 0, 0, 1]])
    cmap_colors = np.concatenate((black, cmap_colors))
    return ListedColormap(cmap_colors)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import LinearSegmentedColormap

# Define the starting color as black and then some colors for the gradient.
# This example uses a simple gradient from black to red to green to blue.
# You can customize this list with any number of colors.
colors = ["#000000", "#ff0000", "#00ff00", "#0000ff"]

# Create the colormap with 2000 colors
cmap = LinearSegmentedColormap.from_list("custom_cmap", colors, N=2000)

# Test the colormap
gradient = np.linspace(0, 1, 256)
gradient = np.vstack((gradient, gradient))

# Plot a test figure
fig, ax = plt.subplots(figsize=(6, 2))
ax.imshow(gradient, aspect='auto', cmap=cmap)
ax.set_axis_off()

plt.show()
```
